### 생각할 거리들
1. 4차 산업혁명
    - 기계가 대체하는 것은 사람의 노동력. 그게 2차산업혁명
    - 이번에는 소프트웨어가 대체하는 영역. 뭐가 다를까?
    - 알고리즘, 소프트웨어는 카피코스트가 제로임. 기계와 달리 하나를 더 만드는데에 대한 비용이 상당히 낮아짐.
    - 포크레인을 유지하는 데에 한달에 100만원의 비용이 든다고 치면, 사람은 99만원에 일하게 됨.
    - 그러면 0에 가까운 소프트웨어의 일을 하던 기존의 노동자의 일자리는 사라짐.
    - 더불어 이건 운반비용이 들지 않음. 운반비용이 없다는 것을 이용해, 아마존이 '가상의 소프트웨어 포크레인'을 만들면 그게 전세계에서 다 쓰일 수 있음.
    - 이 경우, 나라 단위의 문제가 발생함. 후발주자가 따라잡기 힘든, 선두주자에 의한 선점의 문제가 발생함.
    - 그런 상황에서 우리는 어떻게 대처해야 할까?
    
2. 데이터가 갑
    - 데이터분석가가 아무리 뛰어나도, 데이터가 없으면 아무 것도 아니다.
    - 우리나라는 어떻게 데이터를 확보해야 할까?
    
3. 자동화
    - 데이터 직무 자체는 자동화가 되지 않을까?
    - '무엇을 데이터분석할 것인가?'에 대한 고민을 계속 해야 함.
    - '어떤 것에 적용해서, 무엇을 필요로 할 것이며, 어떤 결과를 도출할 것이며, 얼마를 벌 지'에 대해 고민하자.

4. 기본소득
    - 일자리가 대체되었을 때, 알고리즘으로 번 돈에 대한 부가가치는 어떻게 분배되어야 하나?

5. 분석 결과와 암시
    - 같은 결과라도, 분석가가 다는 코멘트 한 줄에 의해 전체가 주는 의미와 인상은 달라질 수 있음.
    - 우리는 분석가로서 어느 정도의 객관성을 유지해야 하나?

6. 데이터와 모델 영향

7. 도메인 지식

8. 다양성과 독립성
    - 소속된 모두가 독립성을 지닌 채로 결정을 내려야 집단의 결정이 좋아질 수 있다.
    - 독립성을 지니려면 내가 스스로에 대한 확신이 있어야 하고, 그를 위한 준비가 되어 있어야 함.
